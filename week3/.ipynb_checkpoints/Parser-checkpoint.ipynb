````python
import requests
from bs4 import BeautifulSoup
import json
import os
````

## Parse headlines from website
### Example usage:
```headlines = parseHeadlines('http://wired.com', '.headline')```

````python
def parseHeadlines(url, domClass):
    if url is not None:
        try:
            response = requests.get(url)
            html = response.text
            soup = BeautifulSoup(html, 'html.parser')
            titles = soup.select(domClass)
            clean_titles = []
            for title in titles:
                clean_titles.append(title.get_text().strip())

            return clean_titles
        except Exception as err:
            print('There was an error')
            print(err)
            
    else:
        print('No URL was provided')
````

````python
headlines = parseHeadlines('https://pornhub.com', '.title')
````

````python
char_list = []

for headline in headlines:
    temp = headline.split(" ")
    for word in temp:
        for char in list(word):
            if char in [" ", "-", "'", '"', "_", '’', '\\u2019', "`", '’']:
                pass
            else:
                char_list.append(char)

char_list_obj = {0: char_list}

with open((os.getcwd() + '/test.json'), 'w') as outfile:
    json.dump(char_list_obj, outfile)
````
